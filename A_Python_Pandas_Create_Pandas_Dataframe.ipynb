# **1. Creating using Constructor**

### **1.1 แนวทางที่ 1 ใช้ Dictionary ทั่วไป**

1 Dict แทน 1 Row

หากต้องการหลาย Row สร้าง List of Dictionary

```python
# 1.1.1 เริ่มจากการสร้างกำหนด List of list ขึ้นมา
technologies = [ ["Spark",20000, "30days"], 
                 ["pandas",20000, "40days"], 
               ]
```

```python
# 1.1.2 นำ List of list จากข้อ 1.1.1 มาใช้กับ Constructor จะได้ Pandas Dataframe ที่มี Default Index Label
# ของทั้ง Row และ Column

import pandas as pd

pd.DataFrame(technologies)
```

```python
# 1.1.3 สร้าง List ขึ้นมาสองตัว ตัวหนึ่งใช้เป็น Row Index Label อีกตัวหนึ่งใช้เป็น Column Index Label
# จากนั้นนำ List ทั้งสองมาใช้เป็น Parameter ของ Constructor ก็จะได้ Pandas Dataframe เป็นผลลัพธ์
# ผลลัพธ์ต่างๆ จาก 1.1.2 ตรงที่ มี Index Label ที่ถูกสร้างขึ้นมาโดย user

row_label=["a","b"]

column_names=["Courses","Fee","Duration"]

pd.DataFrame(technologies,columns=column_names,index=row_label)
```

```python
# 1.1.4 สร้าง List of Dict. (Dict. in List) ขึ้นมาบรรจุ Values และ Column Index Label
# สังเกตว่า มีการใช้ Zip กับ List 2 ตัว ที่มีจำนวน Element เท่ากัน
# column_names ซึ่งเป็น List ที่เก็บ Column Index Label และ row ซึ่งเป็น List ที่เก็บ Values
# dict() เป็น Function ที่ให้เปลี่ยน​โครงสร้างข้อมูลให้เป็น Dictionary
# .append() เป็น Method ที่เพิ่ม Element เข้าไปใน List ซึ่งในที่นี้คือ raw_list 


raw_list = []

for row in technologies:
  raw_list.append(dict(zip(column_names,row)))

raw_list
```

```python
# 1.1.5 นำ raw_list จาก 1.1.4 มาเป็น Parameter ของ Constructor
# สังเกตว่า ผลลัพธ์ที่ได้คือ Pandas Dataframe ที่ไม่มี Row Index Label
pd.DataFrame(raw_list)
```

```python
# 1.1.6 เหมือนกับ 1.1.5 แต่ต่างตรงนี้ มีการเก็บผลลัพธ์ไว้เป็นตัวแปร raw_pd

raw_pd = pd.DataFrame(raw_list)
```

```python
# 1.1.7 .info() เป็น Method ที่สรุปข้อมูลสำคัญของ Dataframe อาทิเช่น Class, Index (Rows), Columns,
# Data Types, Resource Usage เป็นต้น

raw_pd.info()
```

```python
# 1.1.8 .astype() เป็น Method ที่ใช้เปลี่ยน Data Type ของ Columns ใน Dataframe
# สังเกตว่า Format ของ Data Type อยู่ในโครงสร้าง Dictionary

types={'Fee':float}

raw_pd=raw_pd.astype(types)

raw_pd.info()
```

### **1.2 แนวทางที่ 2 ใช้ชุดของ value ใน Dictionary**

มี Dictionary เพียงชิ้นเดียว ไม่มี List of Dictionary เหมือนแนวทางแรก

```python
# 1.2.1 สังเกตว่า มีการกำหนด Row Index Label ด้วย


import numpy as np

technologies   = {
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          }

row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']

pd.DataFrame(technologies, index=row_labels)
```
