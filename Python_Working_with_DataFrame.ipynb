**2.1 การทำ Sub Data**

```python
raw_pd.columns
```

```python
raw_pd[['Name','Sex']]
```

```python
raw_pd.filter(items=['Name','Sex']) # filter columns by name
```

```python
raw_pd.filter(regex='x$',axis=1) # filter columns by regular expression
```

```python
raw_pd['Sex']=='male'
```

```python
type(raw_pd['Sex']=='male')
```

```python
raw_pd[['Sex']]=='male'
```

```python
type(raw_pd[['Sex']]=='male')
```

```python
raw_pd.loc[raw_pd['Sex']=='male'][['Sex']] # filter rows
```

```python
raw_pd.loc[raw_pd['Age']>70] # filter rows
```

```python
raw_pd.query("Sex =='male'") # filter rows
```

```python
raw_pd.loc[raw_pd['Cabin'].isin(['C148'])] # filter rows
```

```python
raw_pd.apply(lambda row: row[raw_pd['Sex']=='male']) # filter rows
```
