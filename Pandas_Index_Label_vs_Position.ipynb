```python
import pandas as pd
```

```python
## กรณีไม่กำหนด Index Label
noIndexLabel_pd = pd.DataFrame({'Province':['กาญจนบุรี','เพชรบุรี','ราชบุรี'],'amount':[80,40,0]})
```

```python
noIndexLabel_pd

## สังเกตว่า ได้รับ Index Label ที่ตรงกับ Index Position
```

```python
# ตรวจสอบ index labels
print("Index Labels:", noIndexLabel_pd.index)

# ตรวจสอบ index positions
print("Index Positions:", range(len(noIndexLabel_pd)))
```

```python
noIndexLabel_pd[1:]
```
