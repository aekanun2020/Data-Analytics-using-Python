การสำรวจข้อมูล ด้วยเทคนิค Univariate Analysis แบบทั่วไป

- ดูสรุปจำนวนค่าที่ไม่ใช่ null (non-null) ของแต่ละ column จาก .info()
- ดู data type ของแต่ละ column จาก .info()
- ดู Scale ของข้อมูลจาก max, min (range) จาก .describe()
- ดูการแจกแจงความถี่ของการเกิดแต่ละค่า ด้วย .groupby().size()
- ทำ pivot เพื่อ summarize ข้อมูล ด้วยฟังก์ชั่นทางสถิติต่างๆ ด้วย .groupby()[].agg()

```python
# Create DataFrame with None/Null to work with examples
import pandas as pd
import numpy as np
technologies   = {
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          }
row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']
df = pd.DataFrame(technologies, index=row_labels)
print(df)
```

```python
df
```

```python
df.shape

## รู้ว่า df มี 8 rows และ 4 columns
```

```python
df.info()

## รู้ว่า df มี 8 rows และ 4 columns
## รู้ว่า Fee, Course มี 7 ค่าที่ไม่ใช่ null (non-null)
## จึงสามารถสรุปได้ว่า Fee มี null อยู่ 1 ตัว และ Course มี null อยู่ 1 ตัว
```

```python
df.describe()

## รู้ว่า ช่วงค่าของ Fee คือ 22,000 - 25,000 และ Discount คือ 1,000 - 2,500
```

```python
df.describe().transpose()

## เห็นเช่นเดียวกับก่อนหน้านี้ แต่อาจจะอ่านผลง่ายกว่า
```

```python
df.groupby('Courses').size()

## เห็นว่ามี Courses "Hadoop" จำนวน 1 หลักสูตร ซึ่งต่างจาก "Spark" ที่มี 2 หลักสูตร
## รวมถึงมี Courses อื่นๆ อีก
```

```python
df.groupby('Courses')['Fee'].mean()

## สรุป ค่า mean ของ Fee ในแต่ละ Courses
```

```python
df.groupby('Courses').agg({'Fee': ['mean', 'sum', 'max']})

## สรุป ค่า mean, sum, max ของ Fee ในแต่ละ Courses
```

```python
df.pivot_table(values='Fee', index='Courses', aggfunc='sum', margins=True)

## สรุป ค่า sum ของ Fee ในแต่ละ Courses และมี Total (All) ในตอนท้าย
```
