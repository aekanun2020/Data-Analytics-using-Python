```python
# 1.1.1 เริ่มจากการสร้างกำหนด List of list ขึ้นมา
technologies = [ ["Spark",20000, "30days"],
                 ["pandas",20000, "40days"],
               ]
```

```python
# 1.1.2 นำ List of list จากข้อ 1.1.1 มาใช้กับ Constructor จะได้ Pandas Dataframe ที่มี Default Index Label
# ของทั้ง Row และ Column

import pandas as pd

pd.DataFrame(technologies)
```

```python
test_pd = pd.DataFrame(technologies)

display(test_pd[[1]]) ## subdata with column label index

print("\n")

display(test_pd[0:1]) ## subdata with row label index

print("\n")

display(test_pd[[1]][0:1]) ## subdata with column label index and row label index

print("\n")

display(test_pd[[1,0]]) ## subdata with list of column label index

print("\n")

display(test_pd[0]=='pandas') ## boolean index

print("\n")

display(test_pd[test_pd[0]=='pandas']) ## subdata with boolean index

print("\n")

display(test_pd[[1,0]][test_pd[0]=='pandas']) ## subdata with column label index and boolean index


## การทำ subdata จะใช้ [] ชุดเดียว หรือ 2 ชุด [][] ก็ได้
## หากใช้ชุดเดียว ในโครงสร้าง [ ] หมายถึง sub column แต่หากเป็นโครงสร้าง [ : ] หมายถึง sub row
## การระบุ col เดียวใช้ [col index 1] หากระบุหลาย col ใช้ [[col index 1, col index 2]] หมายเหตุ col เดียวก็ใช้ [[col index 1]] ได้ด้วย
## การระบุ row ใช้ [row index : จำนวน row ซึ่งจะระบุจำนวนหรือไม่ก็ได้]
## index หมายถึง label หรือ position index กรณี column ใช้ column label index เท่านั้น กรณี row ใช้ row label หรือ position index ก็ได้
```

```python
# 1.1.3 สร้าง List ขึ้นมาสองตัว ตัวหนึ่งใช้เป็น Row Index Label อีกตัวหนึ่งใช้เป็น Column Index Label
# จากนั้นนำ List ทั้งสองมาใช้เป็น Parameter ของ Constructor ก็จะได้ Pandas Dataframe เป็นผลลัพธ์
# ผลลัพธ์ต่างๆ จาก 1.1.2 ตรงที่ มี Index Label ที่ถูกสร้างขึ้นมาโดย user

row_label=["a","b"]

column_names=["Courses","Fee","Duration"]

pd.DataFrame(technologies,columns=column_names,index=row_label)
```
