Null Removal เป็นการ ซ่อนหรือลบทิ้ง (Ignorance)

(1) Ignorance ทั้ง Row ที่มี Null อยู่

(2) Ignorance ทั้ง Column ที่มี Null อยู่

(3) จัดเก็บผลการ Ignorance ไว้ในตัวแปร และ Recheck

```python
# Create DataFrame with None/Null to work with examples
import pandas as pd
import numpy as np
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          })
row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']
df = pd.DataFrame(technologies, index=row_labels)
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df['Courses'].isnull()]
```

```python
df[df['Fee'].isnull()]
```

```python
df[df.columns[df.isnull().any()]][df.isnull().any(axis=1)] # Locate Null
```

# **1. Rows Ignorance**

```python
# ทิ้ง Row ที่มี Element เป็น Null
# สังเกตว่า Row ที่ r4 และ r5 ถูกทิ้งไปทั้ง Row

df.dropna(axis=0)
```

```python
df.dropna(subset=['Courses'])
```

```python
df.dropna(subset=['Fee'])
```

# **2. Columns Ignorance**

```python
# ทิ้ง Column ที่มี Element เป็น Null
# สังเกตว่า Courses และ Fee ถูกทิ้งไปทั้ง Column

df.dropna(axis=1)
```

```python
df
```

# **3. จัดเก็บผลไว้ในตัวแปร และ Recheck**

```python
df = df.dropna()
```

```python
# กลับมาสำรวจดู Null อีกครั้ง และพบว่าไม่มี Null แล้ว

df[df.columns[df.isnull().any()]][df.isnull().any(axis=1)]
```
